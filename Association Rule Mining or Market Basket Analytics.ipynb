```python
transaction = []
for i in range(0,7501):
    transaction.append([str(dataset.values[i,j]) for j in range(0,20)])
```

# Building Association Rule Mining or Market Basket Analytics
## apyori rule

```python
#!pip install apyori
```

```python
from apyori import apriori
```

```python
rules = apriori(transaction, min_support=0.01, min_confidence=  0.3, min_lift = 1.1,  min_length=2)
```

```python
rules2 = apriori(transaction, min_support=0.01, min_confidence=  0.3, min_lift = 3,  min_length=2)
```

```python
"""
Keyword arguments:
    min_support -- The minimum support of relations (float).
    min_confidence -- The minimum confidence of relations (float).
    min_lift -- The minimum lift of relations (float).
    max_length -- The maximum length of the relation (integer).
"""
```
